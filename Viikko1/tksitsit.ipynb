```python
import matplotlib.pyplot as plt
```

```python
vuodet = [2014, 2015, 2016, 2017, 2018]
osallistujat = [131,142,124,156,99 + 89]
```

```python
colors = ["b", "b", "b", "b", "r"]
```

```python
ax = plt.bar(vuodet, osallistujat, color = colors)
plt.ylim(ymin = 0, ymax = 205)
plt.yticks()
plt.xlabel("Vuosi")
plt.ylabel("Osallistujat")
plt.title("Teekkarikulttuurisitseille ilmoittautuneet \n 2014-2018")


palkit = ax.patches

for palkki in palkit:
    y_value = palkki.get_height()
    x_value = palkki.get_x() + palkki.get_width() / 2
    
    plt.annotate(y_value, (x_value, y_value), xytext = (0, 12), textcoords = "offset points", ha = "center", va = "top")
```
