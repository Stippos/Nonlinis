### Exercise 1.1. Extreme points of a univariate function

We can find the stationary points of function $f$ by finding the roots of it's derivative.
$$
\begin{align}
f(x) &= 2x^4 - 5x^3 - x^2\\
f'(x) &= 8x^3 - 15^2 - 2x \\
\end{align}
$$

$$
\begin{align}
f'(x) &= 0 \\
8x^3 - 15x^2 - 2x &= 0 \\
x(8x^2 - 15x - 2) &= 0 \\
\end{align}
$$

We can immedeately see that one root is $x = 0$. The other roots can be solved from.

\begin{align}
8x^2 - 15x - 2 &= 0 \\
x &= \frac{15 \pm \sqrt{15^2 - 4 \cdot 8 \cdot 2}}{2 \cdot 8}\\
x = 2\ &\lor\ x = -\frac{1}{8}
\end{align}

So the stationary points of $f$ are (0, 0), (2, -12), ($-\frac{1}{8}$, $-\frac{51}{2048}$).


#### Let's verify by plotting the functions.

### Exercise 1.2 Extreme points of a bivariate function

The extreme points of a bivariate function can be found by finding where the gradient of the function is 0.

$$
\begin{align}
f(x, y) &= (y - x^2)^2 - x^2\\
\nabla f(x, y) &= 0 \\
\end{align}
$$

$$
\begin{align}
\frac{\partial}{\partial x}  ((y - x^2)^2 - x^2)\mathbf{i} + \frac{\partial}{\partial y}  ((y - x^2)^2 - x^2)\mathbf{j} &= \bar 0 \\
(2(y - x^2)(-2x) - 2x)\mathbf{i} + (2(y - x^2))\mathbf{j} &= \bar 0 \\
(-2x^2 + 2y + 1)(-2x)\mathbf{i} + (2y - 2x^2)\mathbf{j} &= \bar 0
\end{align}
